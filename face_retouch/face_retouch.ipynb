```python
import matplotlib.pyplot as plt
from PIL import Image, ImageEnhance
from PIL import ImageFilter
```

```python
img_path = "C:/encore/download/preprocessing/data/CelebAMask-HQ/CelebA-HQ-img/"
```

```python
img_type = ".jpg"
```

```python
img = Image.open(img_path + str(57) + img_type)
```

```python
def face_retouch(img):

    # 밝게
    enhancer = ImageEnhance.Brightness(img) 
    img_light = enhancer.enhance(1.45)

    # Blur
    image_BLUR = img_light.filter(ImageFilter.BLUR)

    # 또렷하게
    image_DETAIL = image_BLUR.filter(ImageFilter.DETAIL) 
    
    # 이미지 여러개 띄우기
    fig = plt.figure()
    rows = 1; cols = 2

    ax1 = fig.add_subplot(rows, cols, 1)
    ax1.set_title('original face')
    ax1.imshow(img)

    ax2 = fig.add_subplot(rows, cols, 2)
    ax2.set_title('retouched face')
    ax2.imshow(image_DETAIL)
```

```python
face_retouch(img)
```

### 여러 이미지 변환

```python
for i in range(0,10):
    img = Image.open(img_path + str(i) + img_type)
    face_retouch(img)    
```
